```python
# funciÃ³n para generar diccionario con formato adecuado (esta es la primera versiÃ³n en 
# donde se eliminan las etiquetas full)
def labels_exportation(csv_final):
    index = 0
    dic = {}
    alertas = []
    while index < len(csv_final):
        if csv_final.loc[index]['label_name'] != 'nada' and \
        csv_final.loc[index]['label_name'] != 'mala' and \
        csv_final.loc[index]['label_name'] != 'full':
            name = str(csv_final.loc[index]['image_name'])
            # todas las etiquetas de una imagen respectiva
            df = csv_final[csv_final.image_name == name]
            
            # Inicializar diccionario
            result = {}
            
            # Rellenar diccionario
            result['image_id'] = name
            image_height = float(csv_final.loc[index]['image_height'])
            image_width = float(csv_final.loc[index]['image_width'])
            #result['height'] = csv_final.loc[index]['image_height']
            #result['width'] = csv_final.loc[index]['image_width']
            result['height'] = image_height
            result['width'] = image_width
            
            # obs1: si no es una foto mala entonces se itera
            # obs2: clasificaciÃ³n multiple -> etiquetas-> bolsa = 0
            #                                             caja = 1
            #                                             otro = 2
            #                                             full = 3
            # obs3: una clase -> etiqueta = 1 (por producto)
            
            boxes = np.empty((0,4))
            area_list = []
            label_list = []
            alerta = 0
            
            for row in df.index:
                # Coordenadas
                x_min = float(df.loc[row]['x_min'])
                y_min = float(df.loc[row]['y_min'])
                bbox_height = float(df.loc[row]['bbox_height'])
                bbox_width = float(df.loc[row]['bbox_width'])
                x_max = bbox_width + x_min
                y_max = bbox_height + y_min
                # comprobaciÃ³n valores de x e y
                if x_min >= x_max or y_min >= y_max or\
                x_max > image_width or y_max > image_height:
                    alerta+=1
                # ObtenciÃ³n de caja
                new_boxes = np.array([[x_min,y_min,x_max,y_max]])
                boxes = np.concatenate((boxes, new_boxes))
                # Area de cada producto
                area_producto = bbox_width*bbox_height
                area_list.append(area_producto)
                # ObtenciÃ³n de etiqueta
                label = df.loc[row]['label_name']
                #if label != 'full':
                #    label_list.append(1)
                label_list.append(1)
            
            area_array = np.array(area_list)
            label_array = np.array(label_list)
            result['boxes'] = boxes
            result['labels'] = label_array
            result['area'] = area_array
            result['iscrowd'] = np.zeros(len(df))
            
            # Diccionario de diccionarios
            if alerta == 0:
                dic[name] = result
            else:
                alertas.append(name)
        
        elif csv_final.loc[index]['label_name'] == 'nada': 
            name = str(csv_final.loc[index]['image_name'])
            # todas las etiquetas de una imagen respectiva
            df = csv_final[csv_final.image_name == name]
            # Inicializar diccionario
            result = {}
            
            # Rellenar diccionario
            result['image_id'] = name
            result['height'] = 0.0
            result['width'] = 0.0
            
            # obs1: si no es una foto mala entonces se itera
            # obs2: clasificaciÃ³n multiple -> etiquetas-> bolsa = 0
            #                                             caja = 1
            #                                             otro = 2
            #                                             full = 3
            # obs3: una clase -> etiqueta = 1 (por producto)
            area_array = np.array([])
            result['boxes'] = np.empty((0,4))
            result['labels'] = np.ones(0)
            result['area'] = area_array
            result['iscrowd'] = np.zeros(0)
            
            # Diccionario de diccionarios
            dic[name] = result
        # Actualizar indice
        index += len(df)
    return dic, alertas
```
